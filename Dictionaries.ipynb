?2) Create a dictionary of capitals of India and England by list pairing. 
    
     1. Remove the key England and return the value associated with it.




```python
country = ['India','England']
capital = ['New Delhi','London']
country_capital_list = list(zip(country,capital))
country_capital_dict = dict(country_capital_list)
country_capital_dict
```

```python
country_capital_dict.pop('England')
```

**There are nested dictionaries like nested lists**
